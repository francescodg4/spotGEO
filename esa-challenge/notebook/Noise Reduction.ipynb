```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from skimage import filters
from skimage import io


I = np.zeros((50, 50))
I[25,25] = 1.0
I += 0.2*np.random.random(size=(50,50))
    
I = filters.gaussian(I, sigma=0.5)

plt.hist(I.ravel(), log=True)
plt.show()

plt.imshow(I, vmin=0, vmax=1)
plt.show()

plt.imshow(filters.laplace(I), vmin=0, vmax=1)
plt.show()
```

```python
from skimage import exposure

Iin = io.imread('../images/sample2.png', as_gray=True)
Iex = exposure.adjust_sigmoid(Iin, cutoff=0.5)

plt.figure(figsize=(15,5))

plt.imshow(np.hstack([Iin, Iex]), vmin=0, vmax=1)
plt.show()
```

```python
Ilapl = np.absolute(filters.laplace(Iex))

plt.hist(Ilapl.ravel(), log=True)
plt.show()

plt.imshow(Ilapl)
plt.show()
```

```python
Imask = Ilapl > 0.1

plt.imshow(Imask)
plt.show()
```
