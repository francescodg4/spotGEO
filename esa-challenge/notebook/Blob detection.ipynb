```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from skimage import io

I = io.imread('../images/sample.png', as_gray=True)

plt.imshow(I, vmin=0, vmax=1)
plt.show()
```

```python
from skimage import filters

Ilapl = filters.laplace(I)
Ilapl_abs = np.absolute(Ilapl)

plt.hist(Ilapl.ravel())
plt.show()

plt.imshow(np.hstack([Ilapl, Ilapl_abs]))
plt.show()
```
